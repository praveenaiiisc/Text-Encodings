# 2. Stemming And Lemmatization


*   In Stemming, some of the words may not have any meaning, so it not take more time to produce new word(it produces just base word)
*   It is use is sentiment analysis(or classification) bcz not required meaning
full word

---


*   Lemmatization, produces meaningfull word so it will take lot of timing bcz producing meaningful words take timg
*   It use it chatboot, Question Answering bcz this application will required more meaningfull

image.png



```python
from nltk.stem import PorterStemmer # import Stremmer library
from nltk.corpus import stopwords   # in paragraph have words like (of, the, and, ...) doen show the any meaningfull importance value when we solve problem of
                                    # sentiment analysis so stopword remove this type of less important word from my paragraph.

sentences = nltk.sent_tokenize(paragraph) # from tokenizer i have list of sentances
stemmer = PorterStemmer()

# Stemming
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])  # find out list of word from every sentences
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))] # cheack every word in the list of words and removes stopswords
    sentences[i] = ' '.join(words)   # updated sentences
sentences
```

```python
import nltk
from nltk.stem import WordNetLemmatizer # Import Limmatizer Library
from nltk.corpus import stopwords

sentences = nltk.sent_tokenize(paragraph) # Tokenize our paragraph into sentences
lemmatizer = WordNetLemmatizer()

# Lemmatization
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
sentences
```

# 3. BagofWords and TF-IDF

*   BagofWords not contain any sementic meaning
*   TF-IDF contain semantic meaning



```python
# Cleaning the texts
import re  # for regularizer
from nltk.corpus import stopwords # for remove less import words
from nltk.stem.porter import PorterStemmer # for strimming
from nltk.stem import WordNetLemmatizer # for Limmatizing

ps = PorterStemmer()
wordnet=WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)  # paragraph convert into list of sentences
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i]) # Replacing all the thing apart from [a-zA-Z] into spaces in the evry sentences from my paragraph
    review = review.lower() # covert all in the lower letter
    review = review.split() # slpit the perticuler sentences wrt space
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))] # remove all stopwords from my sentences and do stremming
    review = ' '.join(review)
    corpus.append(review)

# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)   # we only take top 1500 frequent word
Bag_repre = cv.fit_transform(corpus).toarray()

# Creating the TF-IDF model
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer()
tf_idf_repre = tv.fit_transform(corpus).toarray()

Bag_repre, tf_idf_repre
```

# 4. Word2Vec

*   Sementic information between different words
*   each word represent 32 or more dimentional vectors instead of single numbers



image.png



```python
from gensim.models import Word2Vec
from nltk.corpus import stopwords
import re

# Preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',paragraph)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)

# Preparing the dataset
sentences = nltk.sent_tokenize(text)

sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

for i in range(len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]


# Training the Word2Vec model
model = Word2Vec(sentences, min_count=1)
# words = model.wv.vocab
words = model.wv.key_to_index
words
```

```python
# Finding Word Vectors
vector = model.wv['war']
vector
```

```python
# Most similar words
similar = model.wv.most_similar('vikram')
similar
```

# 5. Word-Embedding


*   typically in the form of real valued vector that the meaning of the word such that the words that are the closer in the vector space are expected to be similar in meaning
*   image.png




```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
### Vocabulary size
voc_size=10000

### One hot representation
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)
```

```python
# Word_Embedding Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
dim=10
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
model.summary()
```

```python
print(model.predict(embedded_docs))
```
